```python
import time

start_time_temp = []    
for i in range(259622):   
    st = time.strptime(start_time[i], "%Y-%m-%d %H:%M:%S") 
    st_temp = st.tm_min *100 + st.tm_sec
   
    start_time_temp.append(st_temp)
    
start_time_temp
```

```python
import matplotlib.pyplot as plt


X=[0,0,0,0,0,0]
for i in start_time_temp:
    if i <1000:
        X[0] = X[0]+1
    elif i < 2000:
        X[1] = X[1]+1
    elif i < 3000:
        X[2] = X[2]+1
    elif i < 4000:
        X[3] = X[3]+1
    elif i < 5000:
        X[4] = X[4]+1
    else:
        X[5] = X[5]+1

labels=['8:00-8:10','8:10-8:20','8:20-8:30','8:30-8:40','8:40-8:50','8:50-9:00'] 

fig = plt.figure(figsize=(5,5))
plt.pie(X,labels=labels,radius=1,wedgeprops=dict(width=1,edgecolor="w")) 
plt.title("distribution of the start time")
```

```python
end_time_temp = []    
for i in range(259622):   
    ed = time.strptime(end_time[i], "%Y-%m-%d %H:%M:%S") 
    ed_temp = ed.tm_min *100 + ed.tm_sec
   
    end_time_temp.append(ed_temp)
    
end_time_temp
```

```python
import matplotlib.pyplot as plt


X2=[0,0,0,0,0,0]
for i in end_time_temp:
    if i <1000:
        X2[0] = X2[0]+1
    elif i < 2000:
        X2[1] = X2[1]+1
    elif i < 3000:
        X2[2] = X2[2]+1
    elif i < 4000:
        X2[3] = X2[3]+1
    elif i < 5000:
        X2[4] = X2[4]+1
    else:
        X2[5] = X2[5]+1

labels=['8:00-8:10','8:10-8:20','8:20-8:30','8:30-8:40','8:40-8:50','8:50-9:00'] 

fig = plt.figure(figsize=(5,5))
plt.pie(X2,labels=labels,radius=1,wedgeprops=dict(width=1,edgecolor="w")) 
plt.title("distribution of the end time")
```
